```python
# import libraries

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# import dataset

nyc311 = pd.read_csv('311.csv')
```

```python

# info
nyc311.info()
```

```python
# column names
nyc311.columns
```

```python
# shape of the Data frame
nyc311.shape
```

```python
# check unique descriptive values
nyc311['Descriptor'].unique()
```

```python
# check complaint types
nyc311['Complaint Type'].unique()
```

```python
# group the values by complaint type and Descriptor
nyc311.groupby(['Complaint Type','Descriptor']).size()
```

```python
#count complaint types according to city
complaint_type_city= pd.DataFrame({'count':nyc311.groupby(['Complaint Type','City']).size()}).reset_index()
complaint_type_city
```

```python
# major complaints and display graph
nyc311['Complaint Type'].value_counts().head(10).plot(kind='barh',figsize=(5,5));
```

```python
# display major complaint types and count
major_complaint=nyc311.dropna(subset=['Complaint Type'])
major_complaint = nyc311.groupby('Complaint Type')
sorted_complaint_type = major_complaint.size().sort_values(ascending=False)
sorted_complaint_type = sorted_complaint_type.to_frame('count').reset_index()
sorted_complaint_type.head(10)
```

```python
# pie chart with major complaint types
sorted_complaint_type=sorted_complaint_type.head()
plt.figure(figsize=(10,10))
plt.pie(sorted_complaint_type['count'],labels=sorted_complaint_type['Complaint Type'],autopct='%1.1f%%')
plt.show()
```
